```python
import sklearn
import pandas as pd
import numpy as np

df = pd.read_csv('train.csv')
```

```python

y=df['satisfaction']
```

```python
X = df.loc[:, df.columns != 'satisfaction']
X.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
l.fit(X[['Gender']])
df['Gender_res'] = l.transform(df[['Gender']])
df = df.drop('Gender', axis=1)
df
```

```python
l = LabelEncoder()
l.fit(X[['Type of Travel']])
df['Type of Travel_res'] = l.transform(df[['Type of Travel']])
df = df.drop('Type of Travel', axis=1)
df
```

```python

l = LabelEncoder()
l.fit(y)
df['satisfaction_res'] = l.transform(df[['satisfaction']])
df = df.drop('satisfaction', axis=1)
#df
```

```python
l.fit(X[['Customer Type']])
df['Customer Type_res'] = l.transform(df[['Customer Type']])
df = df.drop('Customer Type', axis=1)
df
```

```python
l.fit(X[['Class']])
df['Class_res'] = l.transform(df[['Class']])
df = df.drop('Class', axis=1)
df
```

```python
df = df.drop('id', axis=1)
df
```

```python
y=df['satisfaction_res']
X_ = df.loc[:, df.columns != 'satisfaction_res']
X_.isnull().sum()

```

```python
X = X_.dropna(axis=1)
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X_train , X_test, y_train , y_test = train_test_split(X, y, test_size = 0.2)
clf = DecisionTreeClassifier()
clf.fit(X_train , y_train)
y_predict = clf.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_predict,y_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(solver='liblinear').fit(X_train,y_train)
clf
```

```python
y_predict = clf.predict(X_test)
y_predict
```

```python
y_predict_prob = clf.predict_proba(X_test)
y_predict_prob
```

```python
print('Accuracy of Logistic regression classifier on train set:' , clf.score(X_train, y_train))
print('Accuracy of Logistic regression classifier on test set:' , clf.score(X_test, y_test))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)


clf = LogisticRegression().fit(X_train_scaled, y_train)

print('Accuracy of Logistic regression classifier on test set:' , clf.score(X_train_scaled, y_train))
print('Accuracy of Logistic regression classifier on test set:' , clf.score(X_test_scaled, y_test))
```

```python
from sklearn.svm import SVC
#from adspy_shared_utilities import plot_class_regions_for_classifier_subplot

this_C = 1.0
clf = SVC(kernel = 'linear', C=this_C).fit(X_train, y_train)
print('Accuracy of Linear SVC classifier on training set: {:.2f}'.format(clf.score(X_train, y_train)))
print('Accuracy of Linear SVC classifier on test set: {:.2f}'.format(clf.score(X_test, y_test)))
```
