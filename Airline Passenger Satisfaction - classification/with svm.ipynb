```python
import sklearn
import pandas as pd
import numpy as np

df = pd.read_csv('train.csv')
```

```python

y=df['satisfaction']
```

```python
X = df.loc[:, df.columns != 'satisfaction']
X.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
l.fit(X[['Gender']])
df['Gender_res'] = l.transform(df[['Gender']])
df = df.drop('Gender', axis=1)
df
```

```python
l = LabelEncoder()
l.fit(X[['Type of Travel']])
df['Type of Travel_res'] = l.transform(df[['Type of Travel']])
df = df.drop('Type of Travel', axis=1)
df
```

```python

l = LabelEncoder()
l.fit(y)
df['satisfaction_res'] = l.transform(df[['satisfaction']])
df = df.drop('satisfaction', axis=1)
#df
```

```python
l.fit(X[['Customer Type']])
df['Customer Type_res'] = l.transform(df[['Customer Type']])
df = df.drop('Customer Type', axis=1)
df
```

```python
l.fit(X[['Class']])
df['Class_res'] = l.transform(df[['Class']])
df = df.drop('Class', axis=1)
df
```

```python
df = df.drop('id', axis=1)
df
```

```python
y=df['satisfaction_res']
X_ = df.loc[:, df.columns != 'satisfaction_res']
X_.isnull().sum()

```

```python
X = X_.dropna(axis=1)
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X_train , X_test, y_train , y_test = train_test_split(X, y, test_size = 0.2)
clf = DecisionTreeClassifier()
clf.fit(X_train , y_train)
y_predict = clf.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_predict,y_test)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier

clf_svm=BaggingClassifier(SVC(kernel = 'rbf',degree=5,probability=True),n_estimators=5,max_samples=2000)
clf_svm.fit(X_train, y_train)

clf_svm.score(X_test,y_test)
```

```python

clf_svm=BaggingClassifier(SVC(kernel = 'rbf',degree=5,C=5,probability=True),n_estimators=5,max_samples=2000)
clf_svm.fit(X_train, y_train)

clf_svm.score(X_test,y_test)
```

```python
clf_svm=BaggingClassifier(SVC(kernel = 'rbf',degree=4,C=5,probability=True),n_estimators=10,max_samples=1000)
clf_svm.fit(X_train, y_train)

clf_svm.score(X_test,y_test)
```

```python
for this_gamma in [0.01, 1, 5]:
    
    for this_C in [0.1, 1, 15, 250]:
        
        
        clf_loop = BaggingClassifier(SVC(kernel = 'rbf', gamma = this_gamma,C = this_C,probability=True),n_estimators=10,max_samples=1000).fit(X_train, y_train)
        print("Accuracy on train when gamma: ",this_gamma,"and C:",this_C,clf_loop.score(X_train, y_train) )
        print("Accuracy on test when gamma: ",this_gamma,"and C:",this_C,clf_loop.score(X_test,y_test) )
        
```
